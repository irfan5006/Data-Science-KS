# **Electric Vehicle Sales by State in India**

# About Dataset
This dataset is valuable for analysts, data scientists, and researchers aiming to understand electric vehicle (EV) adoption trends across India. It is versatile and ideal for geographic market segmentation, trend analysis, and predictive modeling. By offering insights into regional EV sales patterns, the dataset supports strategic decision-making in market planning and infrastructure investment.

The data was meticulously scraped from the Clean Mobility Shift website, and then thoroughly preprocessed to ensure accuracy and relevance. All null values have been removed, and the dataset has been cleaned to prepare it for immediate use in exploration, visualization, and analytical projects. It is particularly valuable for market trend analysis, infrastructure planning, and policy development within the EV sector. The dataset is provided in CSV format and is ready for analysis.

`Included Files:`
EV_Dataset.csv: Contains state-level data on EV sales, including vehicle types and categories, offering a comprehensive view of EV distribution across Indian states.

`Key Features:`
State: Names of Indian states with recorded EV sales data.
Vehicle Type: Classifications of vehicles, such as two-wheelers and four-wheelers.
Vehicle Category: Further classification into segments like commercial and passenger vehicles.
Electric_Vehicle_Sales_Quantity: The number of EVs sold per state, essential for analyzing adoption trends.

# Data Visualization:

```python
# Checking Trend Over Time 
plt.figure(figsize=(8,4))

sns.lineplot(x="Year", y="EV_Sales_Quantity", data=df_ev, markers=2, color="black")
plt.xlabel("Year")

plt.ylabel("Electric Vehicle Sales Quantity")
plt.title("Trend Over Time EV Sales Quantity")
```

# Oberservation:
High Units Sale in 2024 which is 150.
Low Units Sale in 2022 which is 100, and 2020 Almost 30 units .
In 2014 no sales.

```python
# Checking based on Month

plt.figure(figsize=(8,4))

sns.lineplot(x="Month_Name", y="EV_Sales_Quantity", data=df_ev, markers="0", color="green")
plt.xlabel("Month")
plt.ylabel("Electric Vehicle Sales Quantity")
plt.title("Trend Over time Based by Month EV Sales Quantity")
```

```python
# Checking Categories Vise 

plt.figure(figsize=(8,4))

sns.barplot(x="Vehicle_Category", y="EV_Sales_Quantity", data=df_ev, hue="Vehicle_Category",palette="dark")
plt.xlabel("Vehicle Category")
plt.ylabel("Electric Vehicle Sales Quantity")
plt.title("Category Sales")
plt.xticks(rotation=0);
```

# Lets Explore The Dataset Deeply.

```python
# using groupby function.
df_ev_year = df_ev.groupby("Year")["EV_Sales_Quantity"].sum().reset_index()
df_ev_year 
```

```python
# Plots the Data using Bar Graph

plt.figure(figsize=(8,4))

sns.barplot(x="Year", y="EV_Sales_Quantity", data=df_ev_year, palette="dark")
plt.xlabel("Year")
plt.ylabel("Electric Vehicle Sales Quantity")
plt.title("Total Sales Over Years")
```

# Oberservations:
1. Highest Sales in 2024.
2. Lowest Sales in 2024.
3. Sales increased from 2014 to 2023 except 2019.
4. Again Sales decreased from 2024.

```python
# vechicle groupby 
vehicle_class=df_ev.groupby('Vehicle_Type')['EV_Sales_Quantity'].sum().sort_values(ascending=False)[:20].reset_index()
vehicle_class 
```

```python
plt.figure(figsize=(16,3))

sns.barplot(x='Vehicle_Type', y='EV_Sales_Quantity', data=vehicle_class, palette='dark')

plt.xlabel('Vehicle Type')

plt.ylabel('Electric Vehicle Sales Quantity')

plt.title('Top 20 Vehicle Types Sales Quantity')
plt.xticks(rotation=45)
```
