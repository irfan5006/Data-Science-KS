# Q.9
 Sort names by their length using lambda function.

```python
babynames.columns
```

```python
#Hint: focus on 2 columns here i.e., NAME and LENGTH (apply appropriate parameters using lambda)

# Create the Name_Length column
babynames["Name_Length"] = babynames["Name"].apply(lambda x: len(x))

# Sort by both 'Name' and 'Name_Length' using different lambda functions
babynames.sort_values(
    by=["Name", "Name_Length"],
    key=lambda x: x.str.len() if x.name == "Name" else x,
    ascending=[True, False],
    inplace=True
)
babynames
```

```python
#Part 1

#Use .sample() and explain why should/could we use it to display a single row

babynames["Name"].sample(20)
babynames

# Because we see randomly the length of names, we Deal only Name column to check quick overview about the entries.

#Part 2

#Use .sample() and explain why should/could we use it to display a ten rows.

babynames["Name"].sample(20).head(10)
# Because we want to quick overiew of entire entries for data integrity.

#Part 3

# Assuming you have the 'babynames' DataFrame

# Set the random seed for reproducibility (read reproducibility in documentation and check out what random_seed does)
import random

# Set the random seed
random.seed(42)  # You can use any number as the seed

# Or if using NumPy
import numpy as np

np.random.seed(42)

# Assuming babynames is your DataFrame
n_samples = 1000  # Number of bootstrap samples
sample_size = len(babynames)  # Sample size (same as the original dataset)

# Initialize lists to store results
means = []
medians = []

# Perform bootstrapping
for _ in range(n_samples):
    # Generate a bootstrap sample with replacement
    bootstrap_sample = babynames.sample(n=sample_size, replace=True)

    # Perform your statistical analysis on the bootstrap sample
    means.append(bootstrap_sample['Name_Length'].mean())  # Calculate mean
    medians.append(bootstrap_sample['Name_Length'].median())  # Calculate median

# Calculate overall statistics
mean_of_means = sum(means) / len(means)
median_of_medians = sum(medians) / len(medians)

# Output the results
print(f"Mean of Bootstrap Means: {mean_of_means}")
print(f"Mean of Bootstrap Medians: {median_of_medians}")



#Hint: The below link is fun to explore and figure out in more detail what all this is:
#https://stackoverflow.com/questions/31502958/drawing-a-bootstrap-sample-from-a-pandas-dataframe
```
